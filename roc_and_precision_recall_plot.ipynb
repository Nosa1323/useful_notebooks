# Генерация данных

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
t = pd.DataFrame({'y_true':[0]*990 + [1]*10})
```

```python
np.random.seed(42)
t['y_score1'] = np.concatenate([np.random.uniform(0,0.2,size=970), # TN
                                np.random.uniform(0.8,1,size=20), # FP
                                np.random.uniform(0,0.2,size=5), # FN
                                np.random.uniform(0.8,1,size=5)]) # TP
```

```python
np.random.seed(42)
t['y_score2'] = np.concatenate([np.random.uniform(0,0.2,size=850), # TN
                                np.random.uniform(0.8,1,size=140), # FP
                                np.random.uniform(0,0.2,size=2), # FN
                                np.random.uniform(0.8,1,size=8)]) # TP
```

# Один график

```python
from sklearn.metrics import RocCurveDisplay, PrecisionRecallDisplay
from matplotlib import pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(6,6))
RocCurveDisplay.from_predictions(t['y_true'],t['y_score1'],name='score1',ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
PrecisionRecallDisplay.from_predictions(t['y_true'],t['y_score1'],name='score1',ax=ax)
plt.legend(loc="upper right")
```

# Два графика на одном

```python
fig, ax = plt.subplots(figsize=(6,6))
RocCurveDisplay.from_predictions(t['y_true'],t['y_score1'],name='score1',ax=ax)
RocCurveDisplay.from_predictions(t['y_true'],t['y_score2'],name='score2',ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(6,6))
PrecisionRecallDisplay.from_predictions(t['y_true'],t['y_score1'],name='score1',ax=ax)
PrecisionRecallDisplay.from_predictions(t['y_true'],t['y_score2'],name='score2',ax=ax)
plt.legend(loc="upper right")
```
